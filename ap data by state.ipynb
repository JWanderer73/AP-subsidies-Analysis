```python
xlsx_file = "datasets\\ap\\school-report-of-ap-exams-grades-11-12-2022-2023.xlsx"
csv_file = "datasets\\ap\\school-report-of-ap-exams-grades-11-12-2022-2023.csv"

df = pd.read_excel(xlsx_file)

df = df.dropna(axis=0, thresh=2)
df = df.dropna(axis=1, how="all")
df
```

```python
pieces = [df.iloc[ : , 0:1], df.iloc[ : , 6:8:1]]
df2 = pd.concat(pieces, axis=1)
df3 = df2.transpose()
df3 = df3.iloc[:, 1:-2]
df3.iat[0,0] = "YEAR"
labels = df3.iloc[0]

merged = df3.iloc[1:3]
merged.columns = labels
merged
```

```python
merged.to_csv("datasets\\ap\\processed\\MERGED-school-report-of-ap-exams-grades-11-12-2022-2023.csv", index=False)
```
