```python
import pandas as pd
import numpy as np

# ===== LOAD DATA FRESH =====
df = pd.read_csv('/Users/alanadams/Downloads/DS3 project/ap_state_2024.csv')

print("="*50)
print("ORIGINAL DATA")
print("="*50)
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())

# ===== STEP 1: RENAME COLUMNS =====
# Create a mapping of old names to new names
rename_dict = {}

# Go through each column and create clean names
for col in df.columns:
    if 'Unnamed: 0' in str(col):
        rename_dict[col] = 'index_col'
    elif 'STATE' in str(col).upper() and 'EXAM' not in str(col).upper():
        rename_dict[col] = 'state'
    elif 'ENROLLMENT' in str(col).upper():
        rename_dict[col] = 'hs_enrollment'
    elif 'EXAM CHG' in str(col).upper():
        rename_dict[col] = 'exams_per_1000_students'
    elif 'AP EXAMS PER 1000' in str(col).upper():
        rename_dict[col] = 'ap_exams_per_1000'
    elif 'GRADERS' in str(col).upper():
        rename_dict[col] = 'grade_11_12_count'
    elif 'SCORES 3' in str(col).upper():
        rename_dict[col] = 'pct_score_3_or_above'

df.rename(columns=rename_dict, inplace=True)

print("\n" + "="*50)
print("AFTER RENAMING")
print("="*50)
print("Columns:", df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())

# ===== STEP 2: REMOVE HEADER ROWS (SAFE VERSION) =====
# Check if 'state' column exists
if 'state' in df.columns:
    print("\n" + "="*50)
    print("REMOVING HEADER ROWS")
    print("="*50)
    
    # Remove rows where state is 'STATE' or NaN
    initial_rows = len(df)
    df = df[df['state'].astype(str) != 'STATE']
    df = df[df['state'].notna()]
    df = df.reset_index(drop=True)
    
    print(f"Removed {initial_rows - len(df)} rows")
    print("\nFirst 5 rows after cleaning:")
    print(df.head())
else:
    print("\nWARNING: 'state' column not found after renaming!")
    print("Skipping header row removal")

# ===== STEP 3: DROP UNNECESSARY COLUMNS =====
print("\n" + "="*50)
print("DROPPING UNNECESSARY COLUMNS")
print("="*50)

cols_to_drop = ['index_col']
# Also drop any 'Unnamed' columns that weren't renamed
cols_to_drop.extend([col for col in df.columns if 'Unnamed' in str(col)])

for col in cols_to_drop:
    if col in df.columns:
        df = df.drop(columns=[col])
        print(f"Dropped: {col}")

print("\nRemaining columns:", df.columns.tolist())

# ===== STEP 4: INSPECT DATA TYPES =====
print("\n" + "="*50)
print("DATA TYPES")
print("="*50)
print(df.dtypes)

# ===== STEP 5: CHECK FOR MISSING VALUES =====
print("\n" + "="*50)
print("MISSING VALUES")
print("="*50)
missing_counts = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
missing_df = pd.DataFrame({
    'Column': missing_counts.index,
    'Missing Count': missing_counts.values,
    'Missing %': missing_pct.values
})
print(missing_df)

# ===== FINAL RESULT =====
print("\n" + "="*50)
print("CLEANED DATAFRAME")
print("="*50)
print("Shape:", df.shape)
print("\nFirst 10 rows:")
print(df.head(10))
print("\nLast 5 rows:")
print(df.tail())

# Save cleaned version
df.to_csv('ap_state_2024_cleaned.csv', index=False)
print("\n✅ Saved to 'ap_state_2024_cleaned.csv'")
```

```python
import pandas as pd
import numpy as np

# Load fresh
df = pd.read_csv('/Users/alanadams/Downloads/DS3 project/ap_state_2024.csv')

# ===== STEP 1: IDENTIFY THE ACTUAL DATA ROWS =====
# Row 0 = headers
# Row 1 = sub-headers with years (2024 2025)
# Row 2+ = actual state data

# Extract column positions for each metric
state_col = df.columns[0]  # 'Unnamed: 0'
enrollment_col = df.columns[2]  # '11TH & 12TH GRADE ENROLLMENT*'

# ===== STEP 2: BUILD CLEAN 2024 DATAFRAME =====
df_2024 = pd.DataFrame()

# Get states (skip first 2 header rows)
df_2024['state'] = df[state_col].iloc[2:].values

# Get 2024 enrollment 
df_2024['hs_enrollment'] = df[enrollment_col].iloc[2:].values

df_2024['year'] = 2024

# Clean: remove NaN/empty states
df_2024 = df_2024[df_2024['state'].notna()]
df_2024 = df_2024[~df_2024['state'].astype(str).isin(['nan', 'NaN', '', 'STATE'])]
df_2024 = df_2024.reset_index(drop=True)

# ===== STEP 3: BUILD CLEAN 2025 DATAFRAME =====
df_2025 = pd.DataFrame()
df_2025['state'] = df[state_col].iloc[2:].values
df_2025['hs_enrollment'] = df[enrollment_col].iloc[2:].values
df_2025['year'] = 2025

df_2025 = df_2025[df_2025['state'].notna()]
df_2025 = df_2025[~df_2025['state'].astype(str).isin(['nan', 'NaN', '', 'STATE'])]
df_2025 = df_2025.reset_index(drop=True)

# ===== STEP 4: CONCATENATE INTO LONG FORMAT =====
df = pd.concat([df_2024, df_2025], ignore_index=True)
df = df.sort_values(['state', 'year']).reset_index(drop=True)

# ===== STEP 5: CLEAN DATA TYPES =====
# Clean enrollment: remove commas, convert to numeric
df['hs_enrollment'] = df['hs_enrollment'].astype(str).str.replace(',', '').str.strip()
df['hs_enrollment'] = pd.to_numeric(df['hs_enrollment'], errors='coerce')

# Clean state names: strip whitespace
df['state'] = df['state'].astype(str).str.strip()

# ===== STEP 6: ADD MORE COLUMNS (PARSE MULTI-VALUE FIELDS) =====
# Now we need to parse the exam data columns that have multiple values

# ===== STEP 6: ADD MORE COLUMNS (PARSE MULTI-VALUE FIELDS) =====
# Reload original to parse the complex columns
df_orig = pd.read_csv('/Users/alanadams/Downloads/DS3 project/ap_state_2024.csv')

# Get the exam column (has format like "31,027 32,017 58,015 60,120")
exam_col = df_orig.columns[3]  # 'EXAM CHG PER 1000...'
score_col = df_orig.columns[6]  # '% OF SCORES 3 OR ABOVE'

# Extract values for each state (skip header rows)
exam_values = df_orig[exam_col].iloc[2:].values
score_values = df_orig[score_col].iloc[2:].values

# Parse exam values (split by spaces)
exam_2024 = []
exam_2025 = []

for val in exam_values:
    if pd.isna(val) or str(val).strip() == '':
        exam_2024.append(np.nan)
        exam_2025.append(np.nan)
    else:
        parts = str(val).split()
        # Assuming first value is 2024, second is 2025
        exam_2024.append(parts[0] if len(parts) > 0 else np.nan)
        exam_2025.append(parts[1] if len(parts) > 1 else np.nan)

# Parse score values (format like "57.1% 63.7%")
score_2024 = []
score_2025 = []

for val in score_values:
    if pd.isna(val) or str(val).strip() == '':
        score_2024.append(np.nan)
        score_2025.append(np.nan)
    else:
        parts = str(val).replace('%', '').split()
        score_2024.append(parts[0] if len(parts) > 0 else np.nan)
        score_2025.append(parts[1] if len(parts) > 1 else np.nan)

# Add these to the respective year dataframes
df_2024_full = df_2024.copy()
df_2024_full['total_ap_students'] = exam_2024
df_2024_full['pct_score_3_or_above'] = score_2024

df_2025_full = df_2025.copy()
df_2025_full['total_ap_students'] = exam_2025
df_2025_full['pct_score_3_or_above'] = score_2025

# Recombine
df = pd.concat([df_2024_full, df_2025_full], ignore_index=True)
df = df.sort_values(['state', 'year']).reset_index(drop=True)

# Clean numeric columns
df['total_ap_students'] = df['total_ap_students'].astype(str).str.replace(',', '').str.strip()
df['total_ap_students'] = pd.to_numeric(df['total_ap_students'], errors='coerce')

df['pct_score_3_or_above'] = pd.to_numeric(df['pct_score_3_or_above'], errors='coerce')
```
